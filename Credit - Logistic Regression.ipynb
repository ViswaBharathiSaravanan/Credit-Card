```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
a=pd.read_excel("https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls",header=1)
a.rename(columns={'default payment next month':'default_payment_next_month'},inplace=True)
#To rename a particular name - dataset.rename(columns={'old name':'new name'})
a.head()
```

<H1>EDA</H1>

```python
y=a.default_payment_next_month
print(y)
```

```python
x=a.drop('default_payment_next_month',axis=1)
print(x)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=100)
print(train_x)
print(test_x)
print(train_y)
print(test_y)
```

```python
model=LogisticRegression().fit(train_x,train_y)
```

```python
predicted_values=model.predict(test_x)
print(predicted_values)
```

```python
probability=model.predict_proba(test_x)
print(probability)
```

```python
pd.DataFrame(probability,predicted_values)
```

```python
confusion_matrix(test_y,predicted_values)
```
